```python
import heapq
def kLargest(lst, k):
    
    y = len(lst)
    smallList = lst[:k]
    heapq.heapify(smallList)
    
    
    for i in range(k, y):
        if smallList[0] < lst[i]:
            heapq.heapreplace(smallList, lst[i])
            
    return smallList
```

```python
n=int(input())
lst=list(int(i) for i in input().strip().split(' '))
k=int(input())
ans=kLargest(lst, k)
for i in ans:
    print(i)
```
