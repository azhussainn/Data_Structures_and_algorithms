```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
    

def ll(arr):
    if len(arr)==0:
        return None
    head = Node(arr[0])
    last = head
    for data in arr[1:]:
        last.next = Node(data)
        last = last.next
    return head

def printll(head):
    while head:
        print(head.data, end=' ')
        head = head.next
    print()
    
def findNrec(head, data):
    
    if head == None:
        return -1
    
    if head.data == data:
        return 0
    
    k = findNrec(head.next, data)
    if k != -1:
        return k + 1
    else:
        return k
    
    
from sys import setrecursionlimit
setrecursionlimit(11000)

arr=list(int(i) for i in input().strip().split(' '))
head = ll(arr[:-1])
data=int(input())
print(findNrec(head, data))
```
